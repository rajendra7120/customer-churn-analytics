```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Display plots inside notebook
%matplotlib inline
```

```python
df = pd.read_csv("../data/cleaned/telco_cleaned.csv")

df.head()
```

```python
plt.figure(figsize=(6,4))

df["Churn"].value_counts().plot(kind="bar")

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

```python
#Most customers remain with the company, but a significant portion has churned. Understanding why these customers leave is critical for improving retention and reducing revenue loss.
```

```python
gender_churn = pd.crosstab(df["gender"], df["Churn"])

gender_churn.plot(kind="bar", figsize=(7,5))

plt.title("Gender vs Customer Churn")
plt.xlabel("Gender")
plt.ylabel("Customers")

plt.show()
```

```python
#Churn appears relatively balanced across genders, indicating gender alone is not a strong predictor of customer churn.
```

```python
contract = pd.crosstab(df["Contract"], df["Churn"])

contract.plot(kind="bar", figsize=(8,5))

plt.title("Contract Type vs Churn")

plt.show()
```

```python
#Customers on month-to-month contracts exhibit significantly higher churn compared with customers on one-year and two-year contracts.
```

```python
internet = pd.crosstab(df["InternetService"], df["Churn"])

internet.plot(kind="bar", figsize=(8,5))

plt.title("Internet Service vs Churn")

plt.show()
```

```python
#Customers using fiber optic services appear to churn more frequently, suggesting opportunities to investigate pricing, service quality, or customer satisfaction.
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["MonthlyCharges"], bins=30)

plt.title("Monthly Charges Distribution")
plt.xlabel("Monthly Charges")
plt.ylabel("Customers")

plt.show()
```

```python
#Monthly charges vary considerably. High-cost plans may contribute to increased churn if customers perceive insufficient value.
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["tenure"], bins=30)

plt.title("Customer Tenure Distribution")
plt.xlabel("Months")
plt.ylabel("Customers")

plt.show()
```

```python
#Many customers leave within the first year, highlighting the importance of onboarding and early customer engagement.
```

```python
plt.figure(figsize=(7,5))

df.boxplot(column="MonthlyCharges", by="Churn")

plt.title("Monthly Charges by Churn")
plt.suptitle("")

plt.show()
```

```python
#Customers who churn tend to have higher monthly charges, suggesting pricing or perceived value may influence retention.
```

```python
# Key Findings

# Month-to-month contracts have the highest churn.
# Fiber optic customers show higher churn rates.
# Customers with shorter tenure are more likely to leave.
# Higher monthly charges are associated with increased churn.
# Gender does not significantly impact churn.
```
